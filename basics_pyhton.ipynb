## pandas

```python
#Importing libraries

import pandas as pd 
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt

#Load the iris 
iris = sns.load_dataset('iris')
```

```python
## Exploring the dataset 
print(iris.head(5))

#display the last rows
print(iris.tail(3))

#statistical summary 
print(iris.describe(include='all'))

#check for missing values 
print(iris.isnull().sum())

```

```python
#Selecting data 
#selecting for single column 
species = iris['species']
print(species)

#selecting multiple columns 
characters = iris[['sepal_length',  'sepal_width',  'petal_length',  'petal_width']]
print(characters)

#by rows index
rows = iris.iloc[:4]
print(rows)
```

```python
#filtering data 
long_petals = iris[iris['petal_length'] >5]
print(long_petals)
```

```python
#sorting data 
sorting_data = iris.sort_values(by='petal_length', ascending=False)
print(sorting_data)
```

```python
# Verify column names
iris.head(3)

# Add the sepal_area column
iris['sepal_area'] = iris['sepal_length'] * iris['sepal_width']

# Display the first few rows
print(iris.head())
```

```python
#drop a column 
iris.drop(columns = ['sepal_area'], inplace = True)
print(iris.head())
```

```python
#aggregating anf group by 
avg = iris.groupby('species').mean()
print(avg)
```

```python
#save to csv / excel

iris.to_csv('iris_data.csv', index=False)

iris.to_excel('iris_data.xlsx', index=False)
```

## Titanic dataset 

```python
titanic = sns.load_dataset('titanic')
print(titanic.head(4))
```

```python
#finding the number of passengers for each class 
class_group = titanic.groupby('class').count()
print(class_group)

class_count = titanic['class'].value_counts()
print(class_count)
```

```python
#grouping passengers by their survival status 
avg_age = titanic.groupby('survived')['age'].mean()


print(avg_age)
```

```python
#Visualize the distribution of ages for male and female passengers
plt.scatter(titanic['who'], titanic['age'], alpha=0.4)
plt.xlabel('Gender (Who)')
plt.ylabel('Age')
plt.title('Age Distribution of Male and Female Passengers')

plt.show()
```

```python
# Sample data
x = np.linspace(0, 10, 100)  # Generate 100 points from 0 to 10
y = np.sin(x) + np.random.normal(0, 0.1, 100)  # Create noisy sine wave

# Create a DataFrame
df = pd.DataFrame({'x': x, 'y': y})

# Apply rolling mean (smoothing)
window_size = 3  # Define the window size for smoothing
df['y_smooth'] = df['y'].rolling(window=window_size).mean()  # Calculate rolling mean

# Plot original noisy data
plt.plot(df['x'], df['y'], label='Original Data', alpha=0.5)

# Plot smoothed data
plt.plot(df['x'], df['y_smooth'], label='Smoothed Data', color='blue')

# Add a legend to differentiate the two plots
plt.legend()

# Display the plot
plt.show()
```

```python
 # Sample data
x = np.linspace(-10, 10, 100)  # Generate 100 points between -10 and 10
y = x**2 + np.random.normal(0, 10, 100)  # Create noisy parabola (y = x^2 + noise)

# Create a DataFrame
df = pd.DataFrame({'x': x, 'y': y})

# Apply rolling mean (smoothing)
window_size = 5  # Define the window size for smoothing
df['y_smooth'] = df['y'].rolling(window=window_size).mean()  # Calculate rolling mean

# Plot original noisy data
plt.plot(df['x'], df['y'], label='Original', alpha=0.5)

# Plot smoothed data (rolling mean)
plt.plot(df['x'], df['y_smooth'], label='Smoothed Parabola', color='green')

# Add a legend to differentiate the two plots
plt.legend()

# Display the plot
plt.show()
```

```python
data = sns.load_dataset('diamonds')
data.head()
```

```python
# Create a scatter plot using seaborn
sns.scatterplot(data=data, x='carat', y='price', hue='cut')

# Add title and labels
plt.title('Scatter Plot: Carat vs. Price by Cut')
plt.xlabel('Carat')
plt.ylabel('Price ($)')

# Display the plot
plt.show()
```

```python
# Box plot of diamond price by cut
sns.boxplot(data=data, x='cut', y='price')
plt.title('Box Plot of Diamond Price by Cut')
plt.xlabel('Cut')
plt.ylabel('Price ($)')
plt.show()
```

```python
# Violin plot of diamond price by cut
sns.violinplot(data=data, x='cut', y='price')
plt.title('Violin Plot of Diamond Price by Cut')
plt.xlabel('Cut')
plt.ylabel('Price ($)')
plt.show()
```

```python
# Joint plot for carat and price
sns.jointplot(data=data, x='carat', y='price', kind='scatter')
plt.title('Joint Plot of Carat and Price')
plt.show()
```
