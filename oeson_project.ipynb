```python
#load libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
import warnings
warnings.filterwarnings('ignore')
```

```python
#load the data
df = pd.read_csv("/Users/niyatkahsay/Desktop/udemy_DS/oeson_interniship/daset_assignment/Global_Temp.csv", sep=',', skiprows=1)

# display the first few rows to check the data structure
print(df.head())

df.info()
```

```python
df.replace('***', np.nan, inplace=True)
```

```python
print(df.head())
```

```python
# convert all columns except 'Year' to numeric values 
df.iloc[:, 1:] = df.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')

print(df.head())
df.shape
```

```python
 #check for missing values in the dataset
missing_data = df.isnull().sum()
print(f"Missing Data Count per Column:\n{missing_data}")

# visualize missing data distribution
plt.figure(figsize=(12, 6))
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Data Heatmap")
plt.show()
```

```python
# line plot - global temperature anomalies over time
plt.figure(figsize=(12, 6))
sns.lineplot(x=df['Year'], y=df['J-D'], marker='o', label="Annual Mean")
plt.axhline(y=0, color='r', linestyle='--', label="Baseline (0C)")
plt.title("Global Land-Ocean Temperature Anomalies Over Time")
plt.xlabel("Year")
plt.ylabel("Temperature Anomaly (C)")
plt.legend()
plt.show()
```

The dataset has 145 rows and 19 columns, representing global land-ocean temperature anomalies across various months and seasons. After inspecting the data, we found that there were no missing values except for the D-N and DJF columns, which had some missing data points.

Looking at the temperature trends over time, we can observe the following:

Between 1880 and 1940, temperatures were consistently below the baseline, indicating that the global temperature was cooler than the reference period.
From 1940 to 1980, temperatures stayed close to the baseline, with no significant upward or downward trend. During this period, the global temperature anomaly was more stable.
After 1980, there was a rapid increase in temperature anomalies, suggesting a sharp rise in global temperatures, which aligns with what we know about accelerating global warming in recent decades.
These observations support the notion that climate change is a recent phenomenon, with temperatures staying relatively stable for much of the 20th century but rising significantly in the last few decades.

```python
# seasonal analysis - group data by seasonal averages
seasons = {
    'DJF': df[['Dec', 'Jan', 'Feb']].mean(axis=1),
    'MAM': df[['Mar', 'Apr', 'May']].mean(axis=1),
    'JJA': df[['Jun', 'Jul', 'Aug']].mean(axis=1),
    'SON': df[['Sep', 'Oct', 'Nov']].mean(axis=1),
}

seasonal_df = pd.DataFrame(seasons)
seasonal_df['Year'] = df['Year']

# plot seasonal variations
plt.figure(figsize=(12, 6))
sns.lineplot(data=seasonal_df, x='Year', y='DJF', label='DJF', marker='o')
sns.lineplot(data=seasonal_df, x='Year', y='MAM', label='MAM', marker='o')
sns.lineplot(data=seasonal_df, x='Year', y='JJA', label='JJA', marker='o')
sns.lineplot(data=seasonal_df, x='Year', y='SON', label='SON', marker='o')
plt.axhline(y=0, color='r', linestyle='--', label="Baseline (0C)")
plt.title("Seasonal Temperature Anomalies Over Time")
plt.xlabel("Year")
plt.ylabel("Temperature Anomaly (C)")
plt.legend()
plt.show()
```

```python
# boxplot- monthly temperature anomalies
plt.figure(figsize=(12, 6))
sns.boxplot(data=df.iloc[:, 1:13])
plt.title("Monthly Temperature Anomalies Distribution")
plt.xlabel("Month")
plt.ylabel("Temperature Anomaly (C)")
plt.xticks(rotation=45)
plt.show()
```

```python
# heatmap seasonal trends
plt.figure(figsize=(12, 6))
sns.heatmap(df.iloc[:, 1:13].T, cmap="coolwarm", annot=False)
plt.title("Monthly Temperature Anomalies (Heatmap)")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```

```python
# correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.iloc[:, 1:].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between Monthly Anomalies")
plt.show()
```

Outlier Observations from the Boxplot:
When we look at the boxplots for each month, its clear that most of the months have outliers. However, the months from October to February stand out, as they show temperature anomalies that are above 1C. These higher-than-normal anomalies suggest that during these months, there were some extreme temperature events, with anomalies significantly above the typical range.

This could indicate unusual climate patterns or extreme weather events that occurred in those months, possibly due to factors like volcanic activity or other environmental influences. These outliers are worth investigating further to understand what might have caused such spikes in temperature anomalies.
